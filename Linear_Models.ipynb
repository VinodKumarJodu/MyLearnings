# Computing Theta

## Using Normal Equation

```python
# theta_best = inv(X_b.T*X_b).X_b.T*y
# y (100,1)
# X_b (100,2) 
# X_b.T (2,100)
# np.linalg.inv(X_b.T.dot(X_b)) (2,2)
# dot(X_b.T).dot(y) (2,1)
# np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y) (2,1)

theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

### Make Predictions Using theta_best

```python
X_new = np.array([[0],[3]])
X_new_b = np.c_[np.ones((2,1)),X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

## Batch Gradient Descent

```python
eta = 0.01
n_iterations = 1000
m = len(X_b)
theta = np.random.rand(2,1)
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot((X_b.dot(theta)-y))
    theta = theta - eta * gradients
    print(theta)
```

```python
theta_bgd_path = []
def plot_bgd(theta,eta,bgd_path = None):
    plt.plot(X,y,'b.')
    n_iterations = 1000 
    m = len(X_b)
    for iteration in range(n_iterations):
        y_predict = X_new_b.dot(theta)
        style='b-' if iteration > 10 else 'r--'
        plt.plot(X_new,y_predict,style)
        gradients = 2/m * X_b.T.dot(X_b.dot(theta)-y)
        theta = theta - eta * gradients
        plt.xlabel('$X$',fontsize=15)
        plt.title("eta = {}".format(eta))
        if bgd_path is not None:
            bgd_path.append(theta)        
```

```python
np.random.seed(19)
theta = np.random.rand(2,1)
plt.figure(figsize=(10,4))
plt.subplot(131),plot_bgd(theta,eta=0.01,bgd_path=theta_bgd_path)
plt.ylabel('$y$',rotation=0,fontsize=15)
plt.subplot(132),plot_bgd(theta,eta=0.05)
plt.subplot(133),plot_bgd(theta,eta=0.1)
```

## Stochastic Gradient Descent

```python
t0,t1 = 1,100
def learning_schedule(t):
    return t0/(t + t1)
```

```python
theta_sgd_path = []
def plot_sgd(theta, n_epochs, sgd_path = None):
    m = len(X_b)
    plt.plot(X,y,"b.")
    for epoch in range(n_epochs):
        for i in range(m):
            y_predict = X_new_b.dot(theta)
            if epoch % 50 == 0:
                style = "b-" if i>30 else "r--"
                plt.plot(X_new,y_predict,style)
                plt.title("$eta={}$".format(learning_schedule(epoch*m)))
            random_index = np.random.randint(m)
            xi = X_b[random_index:random_index+1]
            yi = y[random_index:random_index+1]
            gradients = 2*xi.T.dot(xi.dot(theta)-yi)
            eta = learning_schedule(epoch*m+i)
            theta = theta - eta * gradients
            if sgd_path is not None:
                theta_sgd_path.append(theta)       
```

```python
theta = np.random.rand(2,1)
plt.figure(figsize=(10,4))
plt.subplot(131),plot_sgd(theta,n_epochs=1)
plt.subplot(132),plot_sgd(theta,n_epochs=50,sgd_path=theta_sgd_path)
plt.subplot(133),plot_sgd(theta,n_epochs=100)
```

## Mini Batch Gradient Descent

```python
theta_mgd_path = []
t0,t1 = 10,100
def learning_schedule(t):
    return t0/(t+t1)
```

```python
def plot_mgd(theta,n_epochs,mini_batch_size,mgd_path=None):
    m = len(X_b)
    t = 0
    plt.plot(X,y,'b.')
    for epoch in range(n_epochs):
        y_predict = X_new_b.dot(theta)
        plt.plot(X_new,y_predict)
        for i in range(0,m,mini_batch_size):
            t += 1
            xi = X_b[i:i+mini_batch_size]
            yi = y[i:i+mini_batch_size]
            gradients = 2* xi.T.dot(xi.dot(theta)-yi)
            eta = learning_schedule(t)
            theta = theta - eta * gradients
            if mgd_path is not None:
                theta_mgd_path.append(theta)
```

```python
theta = np.random.randn(2,1)
n_epochs = 1000
plot_mgd(theta,n_epochs,mini_batch_size=20,mgd_path=theta_mgd_path)
```

```python
theta_mgd_path
```

```python
theta_mgd_path = []
def plot_mgd(theta, n_epochs,mini_batch_size, mgd_path = None):
    m = len(X_b)
    plt.plot(X,y,"b.")
    for epoch in range(n_epochs):
        for i in range(1,m,mini_batch_size):
            y_predict = X_new_b.dot(theta)
            if epoch % 50 == 0:
                plt.plot(X_new,y_predict)
                plt.title("$eta={}$".format(learning_schedule(epoch*m)))
            xi = X_b[mini_batch_size:mini_batch_size+1]
            yi = y[mini_batch_size:mini_batch_size+1]
            gradients = 2*xi.T.dot(xi.dot(theta)-yi)
            eta = learning_schedule(epoch*m+i)
            theta = theta - eta * gradients
            if mgd_path is not None:
                theta_mgd_path.append(theta)
```

```python
np.random.seed(19)
theta = np.random.rand(2,1)
n_epochs = 1000
plot_mgd(theta,n_epochs,mini_batch_size=15,mgd_path=theta_mgd_path)
```

## Learning Curve of Plain Linear Regression

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model,X,y):
    X_train,X_val,y_train,y_val = train_test_split(X,y,test_size=0.2,random_state=41)
    train_errors,val_errors = [],[]
    for m in range(1,len(X_train)):
        model.fit(X_train[:m],y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m],y_train_predict))
        val_errors.append(mean_squared_error(y_val,y_val_predict))
    plt.plot(np.sqrt(train_errors),'r-+',linewidth=2,label='train')
    plt.plot(np.sqrt(val_errors),'b-',linewidth=3,label='val')
    plt.legend()
    plt.xlabel('Training set Size')
    plt.ylabel('RMSE')        
```

```python
lin_reg = LinearRegression()
plot_learning_curves(lin_reg,X,y)
plt.show()
```

```python
from sklearn.linear_model import Ridge

np.random.seed(19)
m = 20
X = 3 * np.random.rand(m,1)
y = 1 + 0.5*X + np.random.randn(m,1)
X_new = np.linspace(0,3,100).reshape(100,1)
```

```python
def plot_model(model_class,polynomial,alphas,**model_kargs):
    for alpha,style in zip(alphas,('b-','g--','r.')):
        model = model_class(alpha,**model_kargs) if alpha > 0 else LinearRegression()
        if polynomial:
            model = Pipeline([
                ('poly_features',PolynomialFeatures(degree=10,include_bias=False)),
                ('std_scaler',StandardScaler()),
                ('regul_reg',model),                
            ])
        model.fit(X,y)
        y_new_regul = model.predict(X_new)
        lw = 2 if alpha > 0 else 1
        plt.plot(X_new,y_new_regul,linewidth=lw,label=r'$\alpha = {}$'.format(alpha))
        plt.plot(X,y,'b.')
        plt.legend()
        plt.xlabel('X',fontsize=15)
```

```python
plt.figure(figsize=(8,4))
plt.subplot(121)
plot_model(Ridge,polynomial=False,alphas=(0,10,100),random_state=42)
plt.ylabel("$y$",rotation=0,fontsize=18)
plt.subplot(122)
plot_model(Ridge,polynomial=True,alphas=(0,10**-5,1),random_state=42)
plt.show()
```

# Early Stopping

```python
np.random.seed(42)
m = 100
X = 6*np.random.rand(m,1) - 3
y = 2 + X +0.5 * X**2 + np.random.randn(m,1)
```

```python
X_train,X_val,y_train,y_val = train_test_split(X[:50],y[:50].ravel(),test_size=0.2,random_state=42)

poly_scaler = Pipeline([
    ('poly_features',PolynomialFeatures(degree=90,include_bias=False)),
    ('std_scaler',StandardScaler()),    
])

X_train_poly_scaled = poly_scaler.fit_transform(X_train)
X_val_poly_scaled = poly_scaler.fit_transform(X_val)

sgd_reg = SGDRegressor(max_iter=1,
                      penalty=None,
                      eta0= 0.0005,
                      warm_start=True,
                      learning_rate='constant',
                      random_state=19)
n_epochs=500
train_errors,val_errors = [],[]
for epoch in range(n_epochs):
    sgd_reg.fit(X_train_poly_scaled,y_train)
    y_train_predict = sgd_reg.predict(X_train_poly_scaled)
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    train_errors.append(mean_squared_error(y_train,y_train_predict))
    val_errors.append(mean_squared_error(y_val,y_val_predict))
    
best_epoch = np.argmin(val_errors)
best_val_rmse = np.sqrt(val_errors[best_epoch])
```

```python
plt.annotate('Best Model',
            xy = (best_epoch,best_val_rmse),
            xytext = (best_epoch,best_val_rmse+1),
            ha='center',
            arrowprops=dict(facecolor='black',shrink=0.05),
            fontsize=16,
            )
best_val_rmse -= 0.03
plt.plot([0,n_epochs],[best_val_rmse,best_val_rmse],'k:',linewidth=2)
plt.plot(np.sqrt(val_errors),'b-',linewidth=3,label='Validation Set')
plt.plot(np.sqrt(train_errors),'r--',linewidth=2,label='Training Set')
plt.legend(loc='upper right',fontsize=14)
plt.xlabel('Epoch',fontsize=15)
plt.ylabel('RMSE',fontsize=15)
plt.show()
```

## Implementation of Early Stopping

```python
from sklearn.base import clone
sgd_reg = SGDRegressor(max_iter=1,warm_start=True,penalty=None,
                      learning_rate='constant',eta0=0.0005,random_state=42)

min_val_error = float('inf')
best_epoch = None
best_model = None
for epoch in range(1000):
    sgd_reg.fit(X_train_poly_scaled,y_train)
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    val_error = mean_squared_error(y_val,y_val_predict)
    
    if val_error < min_val_error:
        min_val_error = val_error
        best_epoch = epoch
        best_model = clone(sgd_reg)
        last_increase = 0
    else:
        last_increase += 1
        if last_increase > 5:
            break
```
